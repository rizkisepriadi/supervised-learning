```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import chi2
from sklearn import metrics
```

```python
iris = load_iris()
```

```python
iris.keys()
```

```python
iris_data = pd.DataFrame(iris['data'],columns=iris['feature_names'])
iris_data.head()
```

```python
iris_target = pd.DataFrame(data=iris['target'], columns=['target'])
```

```python
data = pd.concat([iris_data, iris_target], axis=1)
```

Feature Selection

```python
from sklearn.feature_selection import SelectKBest
bestfeature = SelectKBest(score_func=chi2, k=4)
fit = bestfeature.fit(iris_data, iris_target)
fit.scores_
```

```python
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(iris_data.columns)
featureScores = pd.concat([dfcolumns,dfscores], axis=1)
featureScores.columns = ['Iris', 'Scores']
print(featureScores.nlargest(4,'Scores'))
```

Terlihat fitur petal length sangat berpengaruh pada data iris 

::::# **K-Nearest Neighbor with Dataset Irish**

```python
x_train, x_test, y_train, y_test = train_test_split(iris_data, iris_target, test_size=0.3,)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=3,weights='distance',metric='euclidean')
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
print('Accuracy:',metrics.accuracy_score(y_test, y_pred))
```

Terlihat hasil akurasi data dalam memprediksi data baru adalah 97.77%
